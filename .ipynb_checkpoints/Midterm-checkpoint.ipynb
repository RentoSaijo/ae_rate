```python
# Import libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data.
sv = pd.read_csv('SV001.csv').drop('Unnamed: 0', axis=1)
ae = pd.read_csv('AE001.csv').drop('Unnamed: 0', axis=1)
```

```python
# Clean
for df in [sv, ae]:
    df["Subject"] = df["Subject"].astype(str)
    df["Site"] = df["Site"].astype(str)

ae["is_serious"] = (ae.get("SeriousAE", "").astype(str).str.upper() == "YES")
ae = ae.drop('SeriousAE', axis=1)
```

```python
# Clean/Summarise

# Highlight occurred planned visits.
sv["_visit_norm"] = (
    sv["Visits"].astype(str).str.strip().str.lower().str.replace(r"[\s_-]+", "", regex=True)
)
_planned_labels = {"visit1", "visit2", "visit3"}
sv["_is_planned"] = sv["_visit_norm"].isin(_planned_labels)

# Determine subject-level visit counts.
sv_visits_per_subj = (
    sv.groupby(["Site", "Subject"], as_index=False)
      .agg(observed_visits=("Visits", "size"),
           planned_visits_occurred=("_is_planned", "sum"))
)

# Determine site-level counts from sv.
sv_site = (
    sv.groupby("Site")
      .agg(subjects_at_site=("Subject", "nunique"),
           total_observed_visits=("Visits", "size"),
           planned_visits_occurred=("_is_planned", "sum"))   # denominator we want
      .reset_index()
)
sv_site["mean_observed_visits"] = np.where(
    sv_site["subjects_at_site"] > 0,
    sv_site["total_observed_visits"] / sv_site["subjects_at_site"],
    np.nan,
)

# Determine site-level counts from ae.
ae_site = (
    ae.groupby("Site")
      .agg(total_AEs=("Subject", "count"),
           serious_AEs=("is_serious", "sum"),
           subjects_with_any_AE=("Subject", "nunique"))
      .reset_index()
)

# Merge site-level counts from both.
site = sv_site.merge(ae_site, on="Site", how="left").fillna({
    "total_AEs": 0,
    "serious_AEs": 0,
    "subjects_with_any_AE": 0
})

# Calculate AE rate.
site["AE_rate_per_planned_visit"] = np.where(
    site["planned_visits_occurred"] > 0,
    site["total_AEs"] / site["planned_visits_occurred"],
    np.nan,
)

site["AE_rate_per_observed_visit"] = np.where(
    site["total_observed_visits"] > 0,
    site["total_AEs"] / site["total_observed_visits"],
    np.nan,
)

site["pct_subjects_with_any_AE"] = np.where(
    site["subjects_at_site"] > 0,
    site["subjects_with_any_AE"] / site["subjects_at_site"],
    np.nan,
)

site["serious_AE_share"] = np.where(
    site["total_AEs"] > 0,
    site["serious_AEs"] / site["total_AEs"],
    np.nan,
)

site = site.sort_values(by="AE_rate_per_planned_visit", ascending=False).reset_index(drop=True)

# Calculate pooled rate.
total_AEs_all = site["total_AEs"].sum()
total_planned_all = site["planned_visits_occurred"].sum()
pooled_rate = total_AEs_all / total_planned_all if total_planned_all > 0 else np.nan

def control_limits(p, n, z=1.96):
    se = np.sqrt(p * (1 - p) / n)
    lower = np.maximum(0, p - z * se)
    upper = np.minimum(1, p + z * se)
    return lower, upper

# AE categories (overall)
cat_overall = (
    ae.groupby("AE_Category")
      .size()
      .reset_index(name="count")
      .sort_values("count", ascending=False)
)
cat_overall["percent"] = cat_overall["count"] / cat_overall["count"].sum()

# Subject-level AE counts, join to visit counts
subj_ae = (
    ae.groupby(["Site", "Subject"], as_index=False)
      .agg(AE_count=("AE_Category", "count"),
           any_serious=("is_serious", "any"))
)
subj = sv_visits_per_subj.merge(subj_ae, on=["Site", "Subject"], how="left").fillna({
    "AE_count": 0,
    "any_serious": False
})
subj["AE_per_visit"] = np.where(
    subj["observed_visits"] > 0,
    subj["AE_count"] / subj["observed_visits"],
    np.nan,
)

# Export updated site summary
site_out_cols = [
    "Site", "subjects_at_site",
    "planned_visits_occurred", "total_observed_visits",
    "total_AEs", "serious_AEs", "subjects_with_any_AE",
    "AE_rate_per_planned_visit", "AE_rate_per_observed_visit",
    "pct_subjects_with_any_AE", "serious_AE_share"
]
site_summary_path = "summary/site_summary.csv"
site[site_out_cols].to_csv(site_summary_path, index=False)
```

```python
# 5) Funnel plot (occurred planned visits as exposure)
if total_planned_all > 0 and np.isfinite(pooled_rate) and site["planned_visits_occurred"].max() > 0:
    plt.figure()
    x = site["planned_visits_occurred"].values
    y = site["AE_rate_per_planned_visit"].values
    plt.scatter(x, y)
    plt.axhline(pooled_rate, linestyle="--")
    xs = np.linspace(max(1, x.min()), x.max(), 200)
    lowers, uppers = [], []
    for n in xs:
        lo, hi = control_limits(pooled_rate, n, z=1.96)
        lowers.append(lo); uppers.append(hi)
    plt.plot(xs, lowers); plt.plot(xs, uppers)
    plt.xlabel("Occurred planned visits at site (Visit1/2/3 only)")
    plt.ylabel("AE rate per occurred planned visit")
    plt.title("Funnel plot: site AE rates vs exposure (95% limits)")
    plt.tight_layout()
    plt.savefig("figures/fig_funnel_planned.png", dpi=150)
    plt.show()
```

```python
# Use occurred planned visits (Visit1/2/3) as denominator
df = site.loc[site["planned_visits_occurred"] > 0, 
              ["Site", "total_AEs", "planned_visits_occurred"]].copy()

# Pooled rate p0
p0 = df["total_AEs"].sum() / df["planned_visits_occurred"].sum()

# Expected counts and z-scores under Binomial(n_i, p0)
df["expected"] = df["planned_visits_occurred"] * p0
df["var"] = df["planned_visits_occurred"] * p0 * (1 - p0)
df = df[df["var"] > 0]  # guard against zero-variance edge cases
df["z"] = (df["total_AEs"] - df["expected"]) / np.sqrt(df["var"])

# 1) Histogram of standardized residuals
plt.figure()
plt.hist(df["z"].dropna(), bins=15, edgecolor="black")
plt.axvline(0, linestyle="--")
plt.axvline(1.96, linestyle=":")
plt.axvline(-1.96, linestyle=":")
plt.xlabel("Standardized Residual (z)")
plt.ylabel("Number of sites")
plt.title("Residuals under Binomial(n_i, p0)")
plt.tight_layout()
plt.savefig("figures/fig_binomial_residuals_hist.png", dpi=150)
plt.show()

# 2) Observed vs expected counts by site
mx = float(df["expected"].max()) if len(df) else 1.0
plt.figure()
plt.scatter(df["expected"], df["total_AEs"])
plt.plot([0, mx], [0, mx], linestyle="--")  # y = x diagonal
plt.xlabel("Expected AE Count per Site (n_i * p0)")
plt.ylabel("Observed AE Count per Site (k_i)")
plt.title("Observed vs. Expected AE Counts")
plt.tight_layout()
plt.savefig("figures/fig_obs_vs_exp_counts.png", dpi=150)
plt.show()

# 3) Histogram of site AE rates (descriptive)
df["rate"] = df["total_AEs"] / df["planned_visits_occurred"]
plt.figure()
plt.hist(df["rate"].dropna(), bins=15, edgecolor="black")
plt.xlabel("Site AE rate (k_i / n_i)")
plt.ylabel("Number of sites")
plt.title("Distribution of site AE rates")
plt.tight_layout()
plt.savefig("figures/fig_site_rate_hist.png", dpi=150)
plt.show()

# 4) Pearson dispersion statistic (prints to console)
X2 = np.sum(((df["total_AEs"] - df["expected"])**2) / df["var"])
dfree = max(len(df) - 1, 1)
dispersion = X2 / dfree
print(f"Pearson dispersion (X2/df): {dispersion:.3f}  [~1 = binomial OK, >>1 = overdispersion]")
```

```python
# 8) Flag potential under-reporters using occurred planned visits

from math import comb

def binom_cdf(k, n, p):
    """Exact CDF P(K <= k) for Binom(n,p) using stable summation."""
    if n <= 0:
        return np.nan
    k = int(k)
    total = 0.0
    for i in range(0, k+1):
        total += comb(n, i) * (p**i) * ((1-p)**(n-i))
    return min(max(total, 0.0), 1.0)

# 1) One-sided p-values under pooled rate
site["p_under"] = [
    binom_cdf(k=int(k), n=int(n), p=float(pooled_rate))
    if (np.isfinite(pooled_rate) and n > 0) else np.nan
    for k, n in zip(site["total_AEs"], site["planned_visits_occurred"])
]

# 2) Benjamini–Hochberg FDR
valid = site["p_under"].notna()
m = valid.sum()
site["p_under_rank"] = np.nan
site.loc[valid, "p_under_rank"] = site.loc[valid, "p_under"].rank(method="first")
site["p_under_fdr"] = np.nan
site.loc[valid, "p_under_fdr"] = site.loc[valid, "p_under"] * m / site.loc[valid, "p_under_rank"]

# Choose your FDR threshold (e.g., 0.10 for screening; 0.05 for stricter)
FDR_ALPHA = 0.01
site["flag_potential_underreport"] = (site["p_under_fdr"] <= FDR_ALPHA)

# 3) OPTIONAL: draw an exact one-sided lower boundary for the funnel at alpha=0.05
ALPHA_LOWER = 0.001
if np.isfinite(pooled_rate) and site["planned_visits_occurred"].max() > 0:
    xs = np.arange(int(site["planned_visits_occurred"].min()), int(site["planned_visits_occurred"].max())+1)
    # For each n, find the largest k such that CDF(k; n, p0) <= ALPHA_LOWER
    k_bounds = []
    for n in xs:
        k_star = 0
        # increment k until we exceed alpha
        while k_star <= n and binom_cdf(k_star, n, pooled_rate) <= ALPHA_LOWER:
            k_star += 1
        # The boundary rate is k_star/n (first k where CDF > alpha), so use (k_star-1)/n
        k_bounds.append(max(0, k_star-1) / n)
    # plot points (existing)
    plt.figure()
    plt.scatter(site["planned_visits_occurred"], site["AE_rate_per_planned_visit"])
    plt.axhline(pooled_rate, linestyle="--")
    plt.plot(xs, k_bounds)  # exact one-sided lower boundary
    plt.xlabel("Occurred planned visits at site (Visit1/2/3 only)")
    plt.ylabel("AE rate per occurred planned visit")
    plt.title("Funnel plot with exact one-sided lower boundary (α = 0.05)")
    plt.tight_layout()
    plt.savefig("figures/fig_funnel_exact_lower.png", dpi=150)
    plt.show()

# Save the updated table with p-values and flags
cols = [
    "Site", "planned_visits_occurred", "total_AEs",
    "AE_rate_per_planned_visit", "p_under", "p_under_fdr", "flag_potential_underreport"
]
site[cols].sort_values("p_under_fdr").to_csv("figures/site_flags_exact_fdr.csv", index=False)
```

```python
# --- 1) Pull the needed columns and compute pooled rate p0 ---
df = site.loc[site["planned_visits_occurred"] > 0,
              ["Site", "planned_visits_occurred", "total_AEs"]].copy()
df = df.rename(columns={"planned_visits_occurred": "n", "total_AEs": "k"})
p0 = df["k"].sum() / df["n"].sum()

# --- 2) Exact one-sided binomial CDF and per-site p-values ---
def binom_cdf(k, n, p):
    """P(K <= k) for K ~ Binom(n, p)."""
    total = 0.0
    for i in range(0, int(k)+1):
        total += comb(int(n), i) * (p**i) * ((1-p)**(int(n)-i))
    return min(max(total, 0.0), 1.0)

alpha = 0.05
df["p_under"] = [binom_cdf(int(k), int(n), float(p0)) for k, n in zip(df["k"], df["n"])]
df["rate"] = df["k"] / df["n"]
df = df.sort_values("n").reset_index(drop=True)

# --- 3) Compute the exact one-sided lower boundary for each n in range ---
n_min, n_max = int(df["n"].min()), int(df["n"].max())
xs = np.arange(n_min, n_max + 1)
boundary = []
for n in xs:
    # find the largest k such that CDF(k; n, p0) <= alpha
    k_star = 0
    while k_star <= n and binom_cdf(k_star, n, p0) <= alpha:
        k_star += 1
    k_thr = max(0, k_star - 1)  # conservative threshold
    boundary.append(k_thr / n)

# --- 4) Plot: points, pooled mean, exact lower boundary ---
plt.figure()
plt.scatter(df["n"], df["rate"])
plt.axhline(p0, linestyle="--")                # pooled mean
plt.plot(xs, boundary)                         # one-sided lower boundary
plt.xlabel("Occurred Planned Visit")
plt.ylabel("AE Rate per Occurred Planned Visit")
plt.title("Site AE rates vs Exposure (1-sided Exact Lower Boundary, α=0.05)")
plt.tight_layout()
plt.savefig("figures/fig_funnel_one_sided_exact.png", dpi=150)
plt.show()

# --- 5) (Optional) print a small flag list at α=0.05 (no multiple-testing adjustment) ---
flagged = df.loc[df["p_under"] <= alpha, ["Site", "n", "k", "rate", "p_under"]]
print("Pooled rate p0 =", round(p0, 4), "| alpha =", alpha)
print("Flagged sites (one-sided exact, no FDR):")
print(flagged.to_string(index=False) if not flagged.empty else "(none)")
```
